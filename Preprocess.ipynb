## Convert the duration

```
import re

durations = [d for d in train['duration']]
durations_in_s = []

for d in durations:
  duration = d
  duration_w = re.search(r"(\d+)w", duration, re.I)
  duration_w = int(duration_w.group(1)) if duration_w else 0
  duration_d = re.search(r"(\d+)d", duration, re.I)
  duration_d = int(duration_d.group(1)) if duration_d else 0
  duration_h = re.search(r"(\d+)h", duration, re.I)
  duration_h = int(duration_h.group(1)) if duration_h else 0
  duration_m = re.search(r"(\d+)m", duration, re.I)
  duration_m = int(duration_m.group(1)) if duration_m else 0
  duration_s = re.search(r"(\d+)s", duration, re.I)
  duration_s = int(duration_s.group(1)) if duration_s else 0
  duration = 0
  duration += duration_w * 7 * 24 * 60 * 60
  duration += duration_d * 24 * 60 * 60
  duration += duration_h * 60 * 60
  duration += duration_m * 60
  duration += duration_s * 1
  durations_in_s.append(duration)

print(durations[:10])
print(durations_in_s[:10])
train['duration'] = durations_in_s
```

## Convert the publishedAt & channelPublishedAt

```
from datetime import datetime, date
```

```
now = datetime.utcnow()
# now = '2021-03-03 08:43:30.482989'
```

```
# Convert published at
published_at = df['publishedAt']
published_at_in_s = []

for t in published_at:
  result = now - datetime.strptime(t, '%Y-%m-%dT%H:%M:%SZ')
  published_at_in_s.append(result.total_seconds())

print(published_at[:10])
print(published_at_in_s[:10])
print(len(published_at_in_s))

train['publishedAt'] = published_at_in_s
```

```
# Convert channel published at
channel_published_at = df['channelPublishedAt']
channel_published_at_in_s = []

for t in channel_published_at:
  result = None
  try:
    result = now - datetime.strptime(t, '%Y-%m-%dT%H:%M:%S.%fZ')
  except:
    result = now - datetime.strptime(t, '%Y-%m-%dT%H:%M:%SZ')
  channel_published_at_in_s.append(result.total_seconds())

print(channel_published_at[:10])
print(channel_published_at_in_s[:10])
print(len(channel_published_at_in_s))

train['channelPublishedAt'] = channel_published_at_in_s
```
